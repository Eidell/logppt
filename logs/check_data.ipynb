```python
dataset = ["Android", "Apache", "BGL", "Hadoop", "HDFS", "HealthApp", "HPC", "Linux", "Mac", "OpenSSH", "OpenStack", "Proxifier", "Spark", "Thunderbird", "Windows", "Zookeeper"]
```

```python
import pandas as pd
import re
```

```python
def get_parameter_list(s, template_regex):
    """
    :param s: log message
    :param template_regex: template regex with <*> indicates parameters
    :return: list of parameters
    """
    # template_regex = re.sub(r"<.{1,5}>", "<*>", template_regex)
    if "<*>" not in template_regex:
        return []
    template_regex = re.sub(r'([^A-Za-z0-9])', r'\\\1', template_regex)
    template_regex = re.sub(r'\\ +', r'\\s+', template_regex)
    template_regex = "^" + template_regex.replace("\<\*\>", "(.*?)") + "$"
    parameter_list = re.findall(template_regex, s)
    parameter_list = parameter_list[0] if parameter_list else ()
    parameter_list = list(parameter_list) if isinstance(parameter_list, tuple) else [parameter_list]
    return parameter_list
```

```python
r = r'[0-9]+(?:\.[0-9]+){3}(:[0-9]+)?'
for d in dataset:
    # print("-----", d, "-----")
    df = pd.read_csv("{0}/{1}_2k.log_structured_corrected.csv".format(d, d))
    examples = set()
    ip_set = set()
    x = False
    for idx, row in df.iterrows():
        # print(row)
        if "<*>:<*>" not in row['EventTemplate']:
            continue
        if re.findall(r, row['Content']):
            x = True
            ip_set.add(row['EventTemplate'])
        else:
            examples.add(row['EventTemplate'])
    if x:
        print("*****", d, "*****")
        print(len(ip_set))
        [print(x) for x in ip_set]
    else:
        print(d)
    if len(examples) > 0:
        print(len(examples))
        [print(x) for x in examples]
```
